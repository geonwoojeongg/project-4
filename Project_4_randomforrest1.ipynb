```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import statsmodels.api as sm
```

```python
season20_21_data = pd.read_csv("../Project_4/2020-21.csv")
season20_21_data.head(10)
```

```python
season20_21_df = pd.DataFrame(season20_21_data)
list(season20_21_df)
```

```python
season20_21_cleaned = season20_21_df[['Date','Time','HomeTeam','AwayTeam','FTHG','FTAG','FTR','HS',
                                       'AS','HST','AST','HF','AF','HC','AC','HY','AY','HR','AR',]]
season20_21_cleaned
```

```python
season20_21_cleaned = season20_21_cleaned.rename(columns = {'FTHG':'HG',
                                                            'FTAG':'AG',
                                                            'FTR':'Winner',
                                                            'HS':'Home Team Shots',
                                                            'AS':'Away Team Shots',
                                                            'HST':'HT Shots on Target',
                                                            'AST':'AT Shots on Target',
                                                            'HF':'HT Fouls',
                                                            'AF':'AT Fouls',
                                                            'HC':'HT Corners',
                                                            'AC':'AT Corners',
                                                            'HY':'HT Yellow Cards',
                                                            'AY':'AT Yellow Cards',
                                                            'HR':'HT Red Cards',
                                                            'AR':'HT Red Cards'})
season20_21_cleaned
```

```python
season20_21_sorted = season20_21_cleaned.sort_values('HomeTeam', ascending=True)
season20_21_sorted
```

```python
def Result(HG, AG):
    if HG > AG:
        return -1
    elif HG == AG:
        return 0
    elif HG < AG:
        return 1
    
season20_21_sorted['Result'] = season20_21_cleaned.apply(lambda row: Result(row.HG, row.AG), axis = 1)
season20_21_sorted
```

```python
home_team_stats = season20_21_sorted.groupby('HomeTeam').sum()
home_team_stats
```

```python
away_team_stats = season20_21_sorted.groupby('AwayTeam').sum()
away_team_stats 
```

```python
home_team_stats['HTAverageGoals'] = home_team_stats['HG'] / 19 
home_team_stats['HTAverageConceded'] = home_team_stats['AG'] / 19 
away_team_stats['ATAverageGoals'] = away_team_stats['AG'] / 19
away_team_stats['ATAverageConceded'] = away_team_stats['HG'] / 19

season20_21_backup = season20_21_sorted
season20_21_sorted

EPL_total_data = pd.DataFrame()
EPL_total_data['Team'] = home_team_stats.index

EPL_total_data = EPL_total_data.merge(home_team_stats['HTAverageGoals'], left_on='Team', right_index=True)
EPL_total_data = EPL_total_data.merge(home_team_stats['HTAverageConceded'], left_on='Team', right_index=True)
EPL_total_data = EPL_total_data.merge(away_team_stats['ATAverageGoals'], left_on='Team', right_index=True)
EPL_total_data = EPL_total_data.merge(away_team_stats['ATAverageConceded'], left_on='Team', right_index=True)

EPL_total_data.set_index("Team", inplace = True)
season_20_21_table= pd.DataFrame(index = EPL_total_data.index)
scaler = MinMaxScaler()
EPL_total_data = pd.DataFrame(scaler.fit_transform(EPL_total_data), columns = EPL_total_data.columns, index = EPL_total_data.index)
EPL_total_data
```

```python
Home_Away = ['HomeTeam', 'AwayTeam']
Att_Defn_data = season20_21_sorted[Home_Away]

Att_Defn_data = Att_Defn_data.merge(EPL_total_data['HTAverageGoals'], left_on='HomeTeam', right_index=True, how='left')
Att_Defn_data = Att_Defn_data.merge(EPL_total_data['HTAverageConceded'], left_on='HomeTeam', right_index=True, how='left') 
Att_Defn_data = Att_Defn_data.merge(EPL_total_data['ATAverageGoals'], left_on='AwayTeam', right_index=True, how='left') 
Att_Defn_data = Att_Defn_data.merge(EPL_total_data['ATAverageConceded'], left_on='AwayTeam', right_index=True, how='left') 

Att_Defn_data['HT_Attacking_Score'] = (Att_Defn_data['HTAverageGoals'] * Att_Defn_data['ATAverageConceded']); 
Att_Defn_data['AT_Attaching_Score'] = (Att_Defn_data['HTAverageConceded'] * Att_Defn_data['ATAverageGoals']);

Att_Defn_data = Att_Defn_data.merge(season20_21_sorted['Result'], left_index = True, right_index = True)

final_rating = Att_Defn_data
Att_Defn_data
```
