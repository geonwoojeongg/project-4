```python
# rollling average will take a group in it's all take  a set of columns taht we want to compute rolling avg
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(3, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["Home Goals", "Away Goals", "Home Team Shots", "Away Team Shots", "HT Shots on Target", "AT Shots on Target"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
merged_rolling = merged.groupby("HomeTeam").apply(lambda x: rolling_averages(x, cols, new_cols))
merged_rolling
```

```python
merged_rolling = merged_rolling.droplevel("HomeTeam")
merged_rolling
```

### Retraining Our ML Model

```python
def make_predictions(data, predictors):
    train = data[data["Date"] < "2022-08-01"]
    test = data[data["Date"] > "2022-08-01"]
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds, ), index = test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(merged_rolling, predictors + new_cols)
```
