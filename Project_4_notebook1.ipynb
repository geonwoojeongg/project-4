```python
import pandas as pd
season20_21_data = pd.read_csv("../Project_4/2020-21.csv")
```

```python
season20_21_df = pd.DataFrame(season20_21_data)
list(season20_21_df)
```

```python
season20_21_cleaned = season20_21_df[['Date','Time','HomeTeam','AwayTeam','FTHG','FTAG','FTR','HS',
                                       'AS','HST','AST','HF','AF','HC','AC','HY','AY','HR','AR',]]
season20_21_cleaned
```

```python
season20_21_cleaned = season20_21_cleaned.rename(columns = {'FTHG':'Home Goals',
                                                            'FTAG':'Away Goals',
                                                            'FTR':'Winner',
                                                            'HS':'Home Team Shots',
                                                            'AS':'Away Team Shots',
                                                            'HST':'HT Shots on Target',
                                                            'AST':'AT Shots on Target',
                                                            'HF':'HT Fouls',
                                                            'AF':'AT Fouls',
                                                            'HC':'HT Corners',
                                                            'AC':'AT Corners',
                                                            'HY':'HT Yellow Cards',
                                                            'AY':'AT Yellow Cards',
                                                            'HR':'HT Red Cards',
                                                            'AR':'HT Red Cards'})
season20_21_cleaned
```

```python
season20_21_sum = season20_21_cleaned.groupby('HomeTeam').sum()
season20_21_sum.head()
```
